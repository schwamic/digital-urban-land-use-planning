# GPT-4o (Extraction)

```python
from utils.image2base64 import image2base64

from dotenv import load_dotenv, find_dotenv
import os

from langchain_openai import ChatOpenAI
from langchain.schema.messages import AIMessage, HumanMessage

load_dotenv(find_dotenv())

model = ChatOpenAI(
    openai_api_key=os.getenv("OPENAI_API_KEY"),
    model_name="gpt-4o",
    model_kwargs={"top_p": 0, "seed": 42},
    temperature=0,
)
```

```python
pz_path = "../data/processed/bpläne/F11-01-PZ.png"
pze_path = "../data/processed/bpläne/F11-01-PZ-Erklärung.png"
base64_pz = image2base64(pz_path)
base64_pze = image2base64(pze_path)

print(base64_pz)
print(base64_pze)
```

```python
msg = model.invoke([
    AIMessage(
        content=[
            {
            "type": "text",
            "text": "Du bist ein Assistent um Textinformationen aus einer Planzeichnung eines Bebauungsplans zu extrahieren."
            }
        ]
    ),
    HumanMessage(
        content=[
        {
            "type": "text",
            "text": "Das Bild veranschaulicht das Schema einer Nutzungsschablone. Fokusiere dich auf diese Textinformationen."
        },
        {
            "type": "image_url",
            "image_url": {
                "url": f"data:image/jpeg;base64,{base64_pze}",
                "detail": "high"
            }
        },
        {
            "type": "text",
            "text": 'Zähle alle Nutzungsschablonen in der vorliegenden Planzeichnung, die innerhalb des räumlichen Geltungsbreichs liegen. Ausgabe im JSON-Format: {"count": <Anzahl>}'
        },
        {
            "type": "image_url",
            "image_url": {
                "url": f"data:image/jpeg;base64,{base64_pz}",
                "detail": "high"
            }
        }
        ]
    ),
])

print(msg.content)
```
