```python
from utils.openai import OpenAI
from utils.parser import Parser
from utils.runner import Runner

instructions = "Du bist ein Assistent um Textinformationen aus einer Planzeichnung eines Bebauungsplans zu extrahieren. Achte auf Vollständigkeit."
ava = OpenAI(instructions)
parser = Parser()
runner = Runner()
```

### P1) Nutzungsschablone

```python
%store -r msg211_f11_nz
ava.clearContext()
ava.addContext([
    {
        "type": "text",
        "text": f"Tabellenerklärung/Schema der Nutzungsschablone: {msg211_f11_nz}"
    },
])
```

```python
# P1.1) Planzeichnung: Nutzungsschablone – WITHOUT BOUNDING BOX
pz_crop_path = "../data/processed/bpläne/2_zeichnung_textteil_getrennt/F11-PZ-crop.png"

instruction = 'Gehe Schritt für Schritt vor: 1. Finde alle Nutzungsschablonen in der vorliegenden Planzeichnung. 2. Gebe die Nutzungsschablonen jeweils im JSON-Format aus: {<Parameter>:<Wert>}'

async def run():
    return await ava.extractTextFromImages(instruction, [pz_crop_path])

await runner.async_consistency_check(run)
```

```python
# P1.2) Planzeichnung: Nutzungsschablone – WITH BOUNDING BOX
pz_bb_crop_path = "../data/processed/bpläne/2_zeichnung_textteil_getrennt/F11-PZ-bb-crop.png"

instruction = 'Gehe Schritt für Schritt vor: 1. Finde alle rot umrandeten Nutzungsschablonen in der vorliegenden Planzeichnung. 2. Gebe die Nutzungsschablonen jeweils im JSON-Format aus: {<Parameter>:<Wert>}'

async def run():
    return await ava.extractTextFromImages(instruction, [pz_bb_crop_path])

await runner.async_consistency_check(run)
```

```python
# P1.3) Planzeichnung: Nutzungsschablone – VISUAL SEARCH + CROPPING
folder_path = "../data/processed/bpläne/2_zeichnung_textteil_getrennt"
img_name = "/F11-PZ-nz-crop"
image_url_prompts = parser.imageseries2prompts(folder_path, img_name)

def run():
    return ava.request([
        {
            "type": "text",
            "text": 'Hier sind die vorkommenden Nutzungsschablonen der Planzeichnung:'
        },
        *image_url_prompts,
        {
            "type": "text",
            "text": 'Gebe die Werte der jeweiligen Nutzungsschablonen strukturiert im JSON-Format aus: {<Parameter>:<Wert>}'
        }
    ])

results = runner.consistency_check(run)
msg222_f11_nz = results[0]
%store msg222_f11_nz
```
